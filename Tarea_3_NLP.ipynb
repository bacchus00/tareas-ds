### **Parte 1: Modelos Lineales (1 ptos)**

Suponga que tiene un dataset de 10.000 documentos etiquetados por 4 categor칤as: pol칤tica, deporte, negocios y otros. 

**Pregunta 1**: Dise침e un modelo lineal capaz de clasificar un documento seg칰n estas categor칤as donde el output sea un vector con una distribuci칩n de probabilidad con la pertenencia a cada clase. 

Especifique: representaci칩n de los documentos de entrada, par치metros del modelo, transformaciones necesarias para obtener la probabilidad de cada etiqueta y funci칩n de p칠rdida escogida. **(0.5 puntos)**

**Respuesta**: 

sea $|V|$ el vocabulario del corpus.

sea $\vec{x}$ el vector de entrada de $|V|$ dimensiones, que corresponde a la representacion BOW de un documento.

sea $W$ una matriz de $|V|\times4$, correspondiente a los pesos.

sea $\vec{b}$ un vector de $4$ dimensiones, este corresponde a los cesgos

la ecuacion que hace la clasificacion del documento es la siguiente:
$\vec{\hat{y}} = softmax(\vec{x}*W+\vec{b})$

$\vec{\hat{y}}$ corresponde al vector de salida de $4$ dimensione, que en cada posicion indica la probabilidad de pertenencia a una de las $4$ categorias de documentos.

Aqui se aplica la funcion softmax para que se forme una distribuicion de probabilidades en el output. 

La funcion de perdida utilizada es:

$L(\vec{\hat{y}},\vec{y}) = -log(\vec{\hat{y}}_{[j]})$
, con $j$ igual a la posicion de la etiqueta correcta del ejemplo evaluado.

Se ocupa esta funcion de perdida debido a que se trata de un problema multi-clase y de hard classification.



**Pregunta 2**: Explique c칩mo funciona el proceso de entrenamiento en este tipo de modelos y su evaluaci칩n. **(0.5 puntos)**

**Respuesta**: 

Para entrenar el modelo primero se inicializa los parametros ($W$ y $\vec{b}$) con valores aleatorios. Luego, iterativamente, se utiliza datos de entrenamiento y se compara el resultado obtenido por el modelo con la etiqueta real del dato usando la funcion de perdida antes definida. Luego, haciendo uso del proceso conocido como decenso de gradiente estocastico, se ajustan los valores de los parametros de manera de a acercarse a la minima perdida. Este proceso se repite de manera que se acerca cada vez mas a un minimo (no hay garantias de que sea global).

**Pregunta 2**: Explique qu칠 es backpropagation. 쮺uales ser칤an los par치metros a evaluar en la red neuronal anterior durante backpropagation? **(0.25 puntos)**

**Respuesta**:
Back propagation es el proceso de realizar decenso de gradiente estocastico de manera eficiente para multiples capas de neuronas. Primero se calcula la funcion de perdida y luego se procede a calcular la derivada. En el caso del modelo anterior se busca derivar la funcion de perdida con respecto a los valores de las matrices $W^i$ de pesos y los vectore $\vec{b}^i$ de sesgo, parametros que se busca ajustar para minimizar el resultado de la loss function.

**Pregunta 3**: Explique los pasos de backpropagation. En la red neuronal anterior: Cuales son las derivadas que debemos calcular para poder obtener $\vec{\delta^l_{[j]}}$ en todas las capas? **(0.25 puntos)**

**Respuesta**:
Una vez hecho el calculo del output al aplicar un vector de input $\vec{x}$ de ejemplo a la red neuronal se calcula la loss function. Luego se comienza calculando las derivadas de la loss function de la capa superior. Para hacer esto se obtiene primero $\vec{\delta}^4$ de la capa superior. Despues se continua calculando $\vec{\delta}$ para las capas inferiores usando la identidad $\vec{\delta}^l_{[j]} = g'(\vec{h}^l_{[j]})\times\Sigma_k(\vec{\delta}^{(l+1)}_{[k]}\times W^{l+1}_{[j,k]})$, de la que se desprende que el valor en una capa $l$ depende de los valores en capa superior, $l+1$. Finalmente para obtener las derivadas se evaluan lo siguiente $\frac{\partial L}{\partial W^{l}_{[i,j]}} = \vec{\delta}^l_{[j]} \times \vec{z}^{(l-1)}_{[i]}$ donde $z$ corresponde a el resultado de aplicar la funcion de activacion al resultado de la capa anterior.

````python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize # Preguntar si podemos usar esto aaaa

def my_tokenizer(dataset):
  text = ""
  for line in dataset:
    text += line
    text += " "
  tokens = word_tokenize(text)
  return list(dict.fromkeys(tokens))

class WordContextMatrix:

  def __init__(self, window_size, dataset, vocab_size):
    # se sugiere agregar un una estructura de datos para guardar las
    # palabras del vocab y para guardar el conteo de coocurrencia   
    list_tokens = []
    for line in dataset:
      list_tokens.append(word_tokenize(line))

    self.corpus = list_tokens
    self.vocab_size = vocab_size
    self.vocab = my_tokenizer(dataset)[0:vocab_size]
    self.window_size = window_size
    

  def add_word_to_vocab(self, word):
    # Le puede ser 칰til considerar un token unk al vocab
    # para palabras fuera del vocab
    self.vocab.append(word)

  def build_matrix(self):
    matriz = np.zeros((self.vocab_size, self.vocab_size),int)
    df = pd.DataFrame(matriz, columns = self.vocab, index = self.vocab)

    for sentence in self.corpus:
      for ind in range(len(sentence)):
        for ind2 in range(len(sentence)):
          if ind2>ind:
            distancia = ind2-ind
          else:
            distancia = ind-ind2
          if distancia != 0 and distancia <= self.window_size:
            if sentence[ind] in self.vocab and sentence[ind2] in self.vocab:
              df[sentence[ind]][sentence[ind2]]+=1
    self.matriz = df
    return self.matriz
      
  def matrix2dict(self):
    # se recomienda transformar la matrix a un diccionario de embedding.
    dic = dict()
    for ind in self.matriz.columns:
      dic2 = dict()
      for ind2 in self.matriz.columns:
        if self.matriz[ind][ind2] != 0:
          dic2[ind2] = self.matriz[ind][ind2]
      dic[ind] = dic2
    self.diccionario = dic
    return self.diccionario
````

````python
# Probando
corpus = [
  "I like deep learning.",
  "I like NLP.",
  "I enjoy flying."
]
vocab_size = 8
M = WordContextMatrix(1, corpus, vocab_size)
print(M.build_matrix())
print(M.matrix2dict())
````
